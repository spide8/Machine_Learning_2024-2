6. Agrega una nueva columna 'precio_por_area' que calcule el precio por metro cuadrado (precio / area). (1 pts)

```python
# tu código aquí
cantidad_soles = dfGrupo12['precio'].astype(str).str.contains('S/', case=False).sum()
cantidad_usd = dfGrupo12['precio'].astype(str).str.contains('$', case=False).sum()

cantidad_soles, cantidad_usd
```

```python
# tu código aquí
def convertir_precio(precio):
    precio = str(precio)
    if '$' in precio:
        return float(precio.replace('$', '').replace(',', '')) * 3.79
    else:
        return float(precio.replace('S/', '').replace(',', ''))

dfGrupo12["conversion_precio"] = dfGrupo12["precio"].apply(convertir_precio)

dfGrupo12["area"] = dfGrupo12["area"].astype(str).str.replace('m2', '').astype(float)

dfGrupo12["precio_por_area_soles"] = dfGrupo12["conversion_precio"] / dfGrupo12["area"] 

#dfGrupo12["conversion_precio_soles"] = dfGrupo12["conversion_precio"].astype(str) + ' soles'

dfGrupo12
```

7. Agrega una columna `moneda` donde se indique la moneda en la que se encuentra el precio. (0.5pts)

```python
# tu código aquí
def identificar_moneda(precio):
    precio = str(precio)
    if '$' in precio:
        return 'Dólares'
    else:
        return 'Soles'

dfGrupo12["moneda"] = dfGrupo12["precio"].apply(identificar_moneda)

dfGrupo12
```

10. A partir del dataset y las columnas, utilizar un modelo de Regresión Linear de ScikitLearn para encontrar el modelo que pueda explicar la correlación entre `precio` y `area`. (4 pts)

Considerando la siguiente ecuación
$Y = X*w + b$

$Y$: `valor del precio`
$X$: `tamaño del departamento en m2` 

$w$ y $b$: coeficientes 


```python
# tu código aquí
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X = dfGrupo12[["area"]]
y = dfGrupo12["conversion_precio"]


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=27)

modelo = LinearRegression()
modelo.fit(X_train, y_train)


w = modelo.coef_
b = modelo.intercept_

predicciones = modelo.predict(X_test)
mse = mean_squared_error(y_test, predicciones)
```

```python
# imprime tus parámetros w y b aquí
print(f"Coeficiente w (pendiente): {w}")
print(f"Intercepto b: {b}")
print(f"Error Cuadrático Medio (MSE): {mse}")
```

### Opiniones
1. De la primera oración podemos rescatar que, la producción de conjuntos de datos para machine learning es un proceso muy ligado al contexto social y técnico en el que se desarrolla. Los datos no son solo un intento de representar la realidad, sino que también la moldean, reforzando ideas y prejuicios ya existentes. Reconocer esta naturaleza es fundamental para prevenir daños y garantizar un desarrollo ético de la inteligencia artificial. Al comprender el proceso de generación de datos, los desarrolladores y usuarios pueden identificar y mitigar riesgos, promoviendo de esta manera, una IA que beneficie a toda la sociedad.

2. Las oraciones 2 y 3 mencionan que los conjuntos de datos de referencia, los cuales se utilizan como puntos de comparación en el desarrollo de modelos de IA, no solo son instrumentos técnicos sino que también son normas que establecen lo que se considera aceptable en cuanto al rendimiento de esos modelos. Estas normas pueden reforzar prejuicios existentes y excluir otras perspectivas. Como consecuencia, los modelos de IA desarrollados y evaluados en base a estos benchmarks pueden repetir e incluso agregar problemas. Por lo tanto, es importante que estas normas se evalúen y actualicen continuamente para garantizar que los modelos de IA sean justos, precisos y que no generen injusticias.

3. Finalmente en la oración 4 nos da a entender que el poder de nombrar y clasificar en el ámbito de la inteligencia artificial, no es algo neutral, sino una forma de poder que influye en cómo entendemos y tratamos a personas y cosas en el mundo digital. Este poder de definición puede reforzar estereotipos, mantener desigualdades y afectar mucho la identidad y el reconocimiento de personas y grupos. Por eso, es muy importante tener un enfoque crítico y reflexivo al desarrollar sistemas de IA, reconociendo que las decisiones sobre cómo clasificar y categorizar tienen grandes implicaciones sociales y políticas.

### Crítica
La aplicación de la IA en el sector inmobiliario peruano presenta un potencial inmenso, especialmente en la valoración de propiedades. Los algoritmos de IA pueden procesar grandes volúmenes de datos históricos y actuales, identificando patrones y tendencias que probablemente se le escaparían a la percepción humana. Esto permitiría generar tasaciones más precisas y objetivas, reduciendo la incertidumbre y agilizando las transacciones. Sin embargo, la viabilidad de estos proyectos depende en gran medida de la calidad y cantidad de los datos disponibles, y adicionalmente, es crucial garantizar la transparencia y auditabilidad de los algoritmos para evitar sesgos y discriminaciones en las valoraciones.

Si se implementa correctamente, la IA puede revolucionar el sector vivienda en Perú. Más allá de las valoraciones, la IA puede optimizar procesos como la búsqueda de propiedades, la gestión de inmuebles y la atención al cliente. No obstante, la implementación de la IA trae riesgos, de los cuales uno es la dependencia excesiva de los sistemas automatizados, lo que podría generar vulnerabilidades ante ciberataques o fallos técnicos. Además, la IA podría generar desigualdades existentes si los datos utilizados para entrenarla están sesgados. Es fundamental desarrollar políticas y regulaciones que garanticen un uso ético y responsable de la IA en el sector vivienda.
